## Questions
Suppose you're on a game show, and you're given the choice of three doors: behind one door is a car; behind the
others, goats. You pick a door, say No. 1, and the host opens another door, say No. 3, which has a goat. He then says to you, "Do you want to pick door No. 2?" Is it to your advantage to switch your choice?\
Q1) What is the probability to win switching door, knowing that the host knows where it is the price and open always
the door with a goat?\
Q2) What is the probability to win switching door, knowing that the host choose randomly which door to open?\
Q3) What is the probability to win switching door, knowing that the host choose sometimes randomly and sometimes
only goats, which door to open?\
Q4) Can you devise a simulation to test your answers for questions Q1, Q2 and Q3?

## Introduction
Every final question starts exactly asking the same thing: what is the probability to win switching door, under different knowledges? This hilights the nature of the statistics we're going to use, based on statements and not on the outcome of random variables. Let's proceed with a first analysis .\
At the beginning I have three doors, with one car and two goats behind them, in front of me. The probabiltity of the car to be under one particular door is equal for every door. I other words, if I consider the three statements
$$\begin{gather}
C_1\,=\,\,"\textit{The car is behind the door n. 1}"\\
C_2\,=\,\,"\textit{The car is behind the door n. 2}"\\
C_3\,=\,\,"\textit{The car is behind the door n. 3}"
\end{gather}\,,$$
obviously the following probabilities are equals:
$$\mathcal{P}(C_1)=\mathcal{P}(C_2)=\mathcal{P}(C_3)=\frac{1}{3}\,,$$
where 1 is the number of good outcomes (the door picked) 3 is the total number of outcomes (total number of doors). Supposing I pick the door n.1, this translates into the statement
$$I_1\,=\,\,"\textit{I choose the door n. 1}".$$
This anyway doesn't change the probabilities seen before because my choise can't affect the position of the car in any way:
$$\mathcal{P}(C_i|I_j)=\mathcal{P}(C_i)=\frac{1}{3}\,,\qquad\text{with}\quad i,j=1,2,3\,.$$
Opening the door n.3, the host gives me the new information:
$$O_3\,=\,\,"\textit{The door n. 3 is open}"\,.$$
Now the question is: given that I've initially chosen the door n.1 and that the door n. 3 is now open, which is the probability to win changing door (or in other words, that the car is behind the door n. 2)? Which is the value for the following probability:
$$\mathcal{P}(C_2|O_3,I_1)=\,?$$
To calculate it, we have to apply the product rule starting from $\mathcal{P}(C_2,O_3|I_1)$, wich gives us
$$\mathcal{P}(C_2,O_3|I_1)=\mathcal{P}(C_2|I_1)\,\mathcal{P}(O_3|C_2,I_1)=\mathcal{P}(O_3|I_1)\,\mathcal(C_2|O_3,I_1)\,.$$
This implies the following:
$$\mathcal{P}(C_2|O_3,I_1)=\frac{\mathcal{P}(C_2|I_1)\,\mathcal{P}(O_3|C_2,I_1)}{\mathcal{P}(O_3|I_1)}\,,$$
which means that the probability of the car to be behind the door n. 2 (= to win changing door), given that I've chosen the door n. 1 and that the host opened the door n.3 with a goat behind, is equal to the probability of the car to be behind the door n. 2, given that I've chosen the door n. 1, multiplied by the probability for the door n.3 to be open by the host, given that the car is surely behind the door n.2 and that I've chosen the door n. 1, all divided for the probability for the door to be open by the host, given  that I've chosen the door n. 1.\
Let's analize one term at time:
* $\mathcal{P}(C_2|I_1)$ was already calculated and we found that
$$\mathcal{P}(C_2|I_1)=\mathcal{P}(C_2)=\frac{1}{3}\,;$$
* $\mathcal{P}(O_3|I_1)$ can be calculated via marginalization spanning the $C_i$ statements, which means to calculate (using the product rule in the second equality and the previous motivated rule in the third)
$$\mathcal{P}(O_3|I_1)=\sum_i\mathcal{P}(C_i,O_3|I_1)=\sum_i\mathcal{P}(C_i|I_1)\,\mathcal{P}(O_3|C_i,I_1)=\sum_i\frac{1}{3}\,\mathcal{P}(O_3|C_i,I_1)$$
or, more explicitly,
$$\begin{split}
\mathcal{P}(O_3|I_1)&=\mathcal{P}(C_1)\,\mathcal{P}(O_3|C_1,I_1)+\mathcal{P}(C_2)\,\mathcal{P}(O_3|C_2,I_1)+\mathcal{P}(C_3)\,\mathcal{P}(O_3|C_3,I_1)\\
&=\frac{1}{3}\,\bigl[\mathcal{P}(O_3|C_1,I_1)+\mathcal{P}(O_3|C_2,I_1)+\mathcal{P}(O_3|C_3,I_1)\bigr]
\end{split}\,.$$

For this last term and for $\mathcal{P}(O_3|C_2,I_1)$ is fundamental to enter in the host knowledge to uderstand how he choose between the door n. 2 and the door n. 3.

## Q1
In this particular question we suppose that the host knows where the car is, so when he has to choose which door to open after my choise to pick the n.1, he must exploit his information. This turns out useful for me to calculate both $\mathcal{P}(O_3|I_1)$ and $\mathcal{P}(O_3|C_2,I_1)$.\
* For $\mathcal{P}(O_3|I_1)$ we found that
$$\mathcal{P}(O_3|I_1)=\frac{1}{3}\,\bigl[\mathcal{P}(O_3|C_1,I_1)+\mathcal{P}(O_3|C_2,I_1)+\mathcal{P}(O_3|C_3,I_1)\bigr].$$
As I said, in this calculus one has to take into account that the host knows where the car is and surely he won't open that particular door. So, if the car is behind the door n. 1, he as perfectly the same probability to open the door n. 2 as to open the door n. 3, since no car is behind those two doors. That implies $\mathcal{P}(O_3|C_1,I_1)=1/2$. If instead the car is in one of the two doors left, the host is forced to open the one with the goat. This implies $\mathcal{P}(O_3|C_2,I_1)=1$ and $\mathcal{P}(O_3|C_3,I_1)=0$. As a conclusion,
$$\mathcal{P}(O_3|I_1)=\frac{1}{3}\Bigl[\frac{1}{2}+1+0\Bigr]=\frac{1}{2}\,.$$
* Aving understood the assignment, I found in the previous point that
$$\mathcal{P}(O_3|C_2,I_1)=1\,,$$
since the host must open the door n. 3, given that he knows that the car is behind the door n. 2 and that I've chosen the door n. 1.

This concludes that
$$\mathcal{P}(C_2|O_3,I_1)=\frac{\mathcal{P}(C_2|I_1)\,\mathcal{P}(O_3|C_2,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot{1}}{1/2}=\frac{2}{3}\approx66.7\%\,.$$
For completness, the probability to win not changing the initial door is
$$\mathcal{P}(C_1|O_3,I_1)=\frac{\mathcal{P}(C_1|I_1)\,\mathcal{P}(O_3|C_1,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot1/2}{1/2}=\frac{1}{3}\approx33.3\%,$$
exactly half of the probability to win changing the door.

## Q2
If the host doesn't know where the car is, things will change (and will be actually funny if he actually opens the door with the car). Let's see again the two calculations separately.
* $\mathcal{P}(O_3|I_1)$ varyes only conceptually, but not in value, because every choise between the two doors left will be now perfectly random and so
$$\mathcal{P}(O_3|C_1,I_1)=\mathcal{P}(O_3|C_2,I_1)=\mathcal{P}(O_3|C_3,I_1)=\frac{1}{2}\,.$$
This implies that
$$\mathcal{P}(O_3|I_1)=\frac{1}{3}\Bigl[\frac{1}{2}+\frac{1}{2}+\frac{1}{2}\Bigr]=\frac{1}{2}\,.$$
* As I said in the previous point,
$$\mathcal{P}(O_3|C_2,I_1)=\frac{1}{2}\,.$$

So one has, for the probability to win changing the door,
$$\mathcal{P}(C_2|O_3,I_1)=\frac{\mathcal{P}(C_2|I_1)\,\mathcal{P}(O_3|C_2,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot{1/2}}{1/2}=\frac{1}{3}\approx33.3\%\,.$$
For the probability to win not changing the door, this is perfectly equal to Q1:
$$\mathcal{P}(C_1|O_3,I_1)=\frac{\mathcal{P}(C_1|I_1)\,\mathcal{P}(O_3|C_1,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot1/2}{1/2}=\frac{1}{3}\approx33.3\%,$$
Knowing that the host doesn't know where is the car, allows the initial uniform distribution to mantain. The new information $O_3$ is, in this case, useless.\
The remaining third of the total probability goes in the probability of the car to be behind the door n. 3, or in other words in the probability that the host opens the door with the car behind.

## Q3
In this case we don't know how the host choose and so we have to parametrize that unknown. The discrimination between the two case only enters when the host has to make the choise between doors n. 2 and n. 3. We have to set the missing knowledge precisely at that point.
* This means that we have to operate just in the calculation of $\mathcal{P}(O_3|I_1)$, using the parameter $k\in[0,1]$ as random variable, where $k=0$ when the host doesn't know where the car is and $k=1$ when the host knows perfectly where the car is. This translates into, assuming a linear behaviour of the unknown,
$$\mathcal{P}(O_3|I_1)=\frac{1}{3}\,\bigl[\mathcal{P}(O_3|C_1,I_1)+\mathcal{P}(O_3|C_2,I_1)+\mathcal{P}(O_3|C_3,I_1)\bigr]=\frac{1}{3}\Bigl[\frac{1}{2}+\frac{1+k}{2}+\frac{1-k}{2}\Bigr]=\frac{1}{2}\,.$$
As we can see, $\mathcal{P}(O_3|I_1)$ doesn't care about this, because $\mathcal{P}(O_3|C_2,I_1)+\mathcal{P}(O_3|C_3,I_1)=1$ as always.
* Anyway this means a lot for $\mathcal{P}(O_3|C_2,I_1)$, which now leads to
$$\mathcal{P}(O_3|C_2,I_1)=\frac{1+k}{2}\,.$$

This implies that the probability to win changing the door is
$$\mathcal{P}(C_2|O_3,I_1)=\frac{\mathcal{P}(C_2|I_1)\,\mathcal{P}(O_3|C_2,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot{(1+k)/2}}{1/2}=\frac{1+k}{3}\,.$$
The probability to win not changing the door is instead invariant, since $\mathcal{P}(O_3|C_1,I_1)=1/2$, unaffected by the uncertainty:
$$\mathcal{P}(C_1|O_3,I_1)=\frac{\mathcal{P}(C_1|I_1)\,\mathcal{P}(O_3|C_1,I_1)}{\mathcal{P}(O_3|I_1)}=\frac{1/3\cdot1/2}{1/2}=\frac{1}{3}\approx33.3\%.$$
Here too, the remaining probability goes in the probability of the car to be behind the door n. 3, or in other words in the probability that the host opens the door with the car behind.
