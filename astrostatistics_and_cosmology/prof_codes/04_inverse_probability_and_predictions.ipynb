```julia
#Example 2.6 BK3
import Random: rand, seed!
import Distributions: Binomial, pdf

function example2_6()

  N = 10
  Nurn = 11
  nB = 3
  seed!(18953) # Use fixed seed to see the effect of increasing precision over multiple Nexp
  Nexp = 900000000
    
  #URN = empty((N, length(idx_urn)))
  #Create an empty matrix for the 11 urns
  URN = zeros(Int, N, Nurn)
  
  for M in 1:N
  
    println("M = ", M)
  
    URN[1:M,M+1] .= 1 # 1 = BLACK ball, 0 = WHITE ball
  
  end #for M in 0:10
  
  #DEBUG
  for i in 1:11
    println(URN[:,i])
  end

  pnB_given_N = 0.0
  pu = 1.0/11.0
  for u in 0:10
    f_u = u/10.0
    pnB_given_u = pdf(Binomial(N, f_u), nB) #eq 2.23
    pnB_given_N += pu*pnB_given_u #eq 2.24
  end # for u in 0:10

  println("pnB_given_N = ", pnB_given_N)

  pBp1_given_nB_theo = 0.0
  for u in 0:10
    f_u = u/10.0
    pnB_given_u = pdf(Binomial(N, f_u), nB) #eq 2.23
    pu_given_nB = pu*pnB_given_u/pnB_given_N #eq 2.25/2.26

    #DEBUG
    println("pu_given_nB = ", pu_given_nB)

    pBp1_given_nB_theo += f_u*pu_given_nB #eq 2.31
  end # for u in 0:10

  println("pBp1_given_nB_theo = ", pBp1_given_nB_theo)

  pBp1_and_nB = 0 #Joint probability that Ball N+1 is Black and nB == 3
  
  for i in 1:Nexp

    iurn = rand(1:11,1)
    vectorB = rand(URN[:,iurn],N)

    #DEBUG
    #println("vectorB = ", vectorB)

    if sum(vectorB) == nB #nB ==3 and N == 10

      #println("Valid vector B: ",vectorB)

      #Do another draw Ball_{N+1}
      Ball_Np1 = rand(URN[:,iurn],1)[1]
      #println("Next Ball = ", Ball_Np1)
      if Ball_Np1 == 1 #Black
        pBp1_and_nB += 1
      end #if Ball_Np1 == 1 #Black
    
    end #if sum(vectorB) == 3

  end #for i in 1:Nexp

  pBp1_given_nB = (pBp1_and_nB/Nexp)/pnB_given_N #see pg. 28
  println("pBp1_given_nB = ", pBp1_given_nB)
  println("stdev = ", sqrt(pBp1_given_nB*(1.0-pBp1_given_nB)/Nexp) )
  println("Relative_error = ", (pBp1_given_nB - pBp1_given_nB_theo)/pBp1_given_nB_theo)

end #function example2_6()

example2_6() #run the code
```
